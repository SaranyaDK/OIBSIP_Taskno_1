# Data Visualization

```python
sns.set_style("whitegrid")
Species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
sns.pairplot(iris_df, hue ='Species', height =2)
plt.show()
```

From the above plots, *PetalLengthCm*, *PetalWidthCm* both have maximum separated species.
By inspecting them we can classify the flower easily, but in the *SepalLengthCm*, *SepalWidthCm* both have maximum merged species, so we cant classify by using them easily

Replace the Species with numerical value.
I took *1 for Iris-setosa*, *2 for Iris-versicolor* and *3 for Iris-virginica* 

```python
iris_df['Species'].replace({'Iris-setosa' : 1, 'Iris-versicolor' : 2, 'Iris-virginica' : 3}, inplace = True)
print(iris_df.head(10))
```

```python
print(iris_df.tail(10))
```

From the above visualization we can say that Iris-Setosa is completely separated from the other two types

# Correlation Matrix

# Data preparation

```python
X = iris_df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
Y = iris_df['Species']
```

In the data preparation double brackets indicate dataframe and single bracket indicates the pandas series

```python
print(X)
```

```python
print(Y)
```

# Test-Train-Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3)
```

##Check the rows (same rows)

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

# Fitting the models using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

# Prediction

```python
y_test_predicted = model.predict(X_test)
print(y_test_predicted.shape)
print(type(y_test_predicted))
print(Y_test.shape)
print(y_test_predicted)
print(Y_test)
```

# Find Accuracy

*Accuracy for test models*

```python
from sklearn.metrics import accuracy_score
print("Accuracy is : ",accuracy_score(Y_test,y_test_predicted)*100)
```
